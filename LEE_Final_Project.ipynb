```python
Counter(" ".join(uni.Description).split()).most_common(20)
```

I used the Counter but realized that a lot of filler words were the most common. So I wrote a command that ignored these filler words.

```python
ignore = ['at', 'top','firm','also' ,'focuses' ,'now','lets','into', 'this','the','a','small', 'local','if','in','it','of','or', 'and', 'for', 'is', 'an', 'to', 'on', 'from', 'with', 'by', 'its', 'offers', 'provides', 'company', 'as', 'has', 'that', 'offering']
buzzwords = Counter(" ".join(uni.Description).lower().split())
for word in list(buzzwords):
    if word in ignore:
        del buzzwords[word]
print (buzzwords)
```

I converted the results to a panda data frame. 

```python
buzz = pd.DataFrame.from_dict(buzzwords, orient='index').reset_index()
buzz.columns = ['Word', 'Frequency']
buzz.head()
```

I'm going to take the top 16 ranked words since there are a lot of ties.

```python
buzz = buzz.sort_values(by = ('Frequency'), ascending=False).head(16)
buzz = buzz.reset_index()
buzz.head()
```

```python
buzz.plot(x='Word', y='Frequency', kind='barh', xticks = range(0,20,2),color = 'b', stacked = False, fontsize = 12, figsize = (14,7), alpha = 0.3) 
plt.xlabel('Frequency', fontsize = 14)
plt.ylabel('Word', fontsize = 14)
plt.title('Most Frequently Used Buzz Words', fontsize = 14)
plt.show()
```
